```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Enerji ekseni (y)
energy = np.linspace(0, 1.5, 500)

# Valans bandı: alt enerji düzeyi
valence_band = np.full_like(energy, 0.5)

# İletkenlik bandı: üst enerji düzeyi
conduction_band = np.full_like(energy, 1.2)

# Fermi seviyeleri
fermi_intrinsic = np.full_like(energy, 0.85)
fermi_n_type = np.full_like(energy, 1.05)
fermi_p_type = np.full_like(energy, 0.65)
```

```python
plt.figure(figsize=(8, 6))

# Bantları çiz
plt.plot(energy, valence_band, label='Valence Band', color='blue')
plt.plot(energy, conduction_band, label='Conduction Band', color='red')

# Fermi seviyeleri
plt.plot(energy, fermi_intrinsic, '--', label='Intrinsic Fermi Level', color='green')
plt.plot(energy, fermi_n_type, '--', label='n-type Fermi Level', color='purple')
plt.plot(energy, fermi_p_type, '--', label='p-type Fermi Level', color='orange')

# Ayarlar
plt.title('Fermi Energy Level by Band Structure and Doping Types')
plt.xlabel('Position)')
plt.ylabel('Energy (eV)')
plt.ylim(0, 1.5)
plt.legend()
plt.grid(True)
plt.savefig("grafik.png", dpi=300)
plt.show()
```
